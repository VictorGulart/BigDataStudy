```python
# sales by platform
# sales by year
# sales by genre
# sales by publisher
```

```python
df_platform = df.drop(['Rank', 'Genre', 'Publisher' ], axis=1)
```

```python
g_year_platform = df_platform.groupby(["Year", "Platform"]).sum()
```

```python
g_year_platform.index
```

```python
ps3_x360_sales = g_year_platform.loc[(slice(None), ['PS3', 'X360']), :]
ps3_x360_sales = ps3_x360_sales.reset_index()
ps4_xone_sales = g_year_platform.loc[(slice(None), ['PS4', 'XOne']), :]
ps4_xone_sales = ps4_xone_sales.reset_index()

ps3_x360_sales.groupby("Platform").sum()
```

```python
help(sns.lineplot)
```

```python
colors = {
    "X360": "blue",
    "XOne": "cyan",
    "PS3": "orange",
    "PS4": "peach"
}
sns.lineplot(x='Year', y='NA_Sales', hue="Platform", data = ps3_x360_sales, palette=colors )
sns.lineplot(x='Year', y='NA_Sales', hue="Platform", data = ps4_xone_sales )
```

```python
sns.scatterplot(data=ps3_x360_sales, x="Year", y="NA_Sales", hue="Platform", palette=colors)
sns.scatterplot(data=ps4_xone_sales, x="Year", y="NA_Sales", hue="Platform")
```
