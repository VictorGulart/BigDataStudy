```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Numpy example
```

```python
height_in = [74, 74, 72, 72, 73, 69, 69, 71, 76, 71, 73, 73, 74, 74, 69, 70,73, 75, 78, 79]
# weight in pound
weight_lb = [180, 215, 210, 210, 188, 176, 209, 200, 231, 180, 188, 180, 185,160, 180, 185, 189, 185, 219, 230]
```

```python
# convert height from inches to centimeters
height_m = np.array(height_in) * 0.0254
```

```python
weight_kg = np.array(weight_lb) * 0.453592
```

```python
bmi = weight_kg / height_m ** 2
bmi
```

```python
# Creating a new array with boolean values to say the person is light or not when the BMI < 21
light = bmi < 21
light
```

```python
# Now we can find all the people in the bmi array that have a bmi lower then 21 with the new array
bmi[light] # in this case returns only one
```

```python
pd.Series(bmi).describe() # statistical 
```

```python
# Quick scatter plot
plt.scatter(weight_kg, height_m)
plt.title("Weight in Kg and Height in m")
plt.xlabel("Weight (Kg)")
plt.ylabel("Height (m)")
```

```python
gdp = {'Year':[2018, 2017, 2016, 2015, 2014, 2013, 2012, 2011, 2010, 2009],
                'US':[20.5, 19.5, 18.7, 18.2, 17.5, 16.8, 16.2, 15.5, 15.0, 14.4],
                'China':[13.6, 12.1, 11.1, 11.0, 10.4, 9.6, 8.5, 7.6, 6.1, 5.1],
                'Japan':[5.8, 5.4, 5.2, 5.1, 5.0, 4.9, 4.7, 4.6, 4.5, 4.3]}
```

```python
%matplotlib inline
```

```python
plt.plot(gdp['Year'], gdp['China'], label="China", marker="s") # add label to be used in the legend
plt.plot(gdp['Year'], gdp['US'], label="US", marker="o")
plt.plot(gdp['Year'], gdp['Japan'], label="Japan", marker="^")

# Axis labeling
plt.xlabel("Year")
plt.ylabel("GDP (Trillion US$)")

# Plot Title
plt.title("Top 3 World GDP Leaders", color="red", fontsize=20)

# Add legend
plt.legend()
```
